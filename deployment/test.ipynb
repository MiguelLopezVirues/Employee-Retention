```python
import pandas as pd


# Analyze the columns and their unique values
def get_column_info(dataframe):
    column_info = {}
    for column in dataframe.columns:
        if dataframe[column].dtype == 'object' or dataframe[column].nunique() < 20:
            # Categorical columns (or low-cardinality numerical columns)
            column_info[column] = list(dataframe[column].dropna().unique())
        else:
            # Numerical columns
            column_info[column] = {
                "min": dataframe[column].min(),
                "max": dataframe[column].max(),
                "mean": dataframe[column].mean()
            }
    return column_info

# Get the column information
column_info = get_column_info(X)

# Display the column information
for column, values in column_info.items():
    print(f"{column}: {values}")
```
